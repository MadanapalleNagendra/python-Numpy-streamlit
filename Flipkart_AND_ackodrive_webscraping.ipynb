```python
import requests
import re
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup


# # single page
# url="https://www.flipkart.com/search?q=led%20tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
# page=requests.get(url,headers = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'})
# #page

# soup=BeautifulSoup(page.text)
# #soup.prettify()



titles=[]
Rating=[]
no_of_rating=[]
no_of_reviews=[]
Brand=[]
Display=[]
os=[]
resolution=[]
Lanch_Year=[]
Warranty=[]
rq=[]
total=[]
# multiple pages
for i in range(1,42):
    url=f"https://www.flipkart.com/search?q=led+tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    request_header = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'}
    page=requests.get(url,headers=request_header )
    soup=BeautifulSoup(page.text,"html.parser")



    for i in soup.find_all("div",class_="KzDlHZ"):
        titles.append(i.text)




    


    for i in soup.find_all("div",class_="tUxRFH"):
        rating=i.find("div",class_="XQDdHH")
        if rating:
            Rating.append(rating.text)
        else:
            Rating.append(np.nan)

    

    for i in soup.find_all("div",class_="tUxRFH"):
        rating=i.find("div",class_="XQDdHH")
        t = i.find("span", class_="Wphh3N")
        if t:  
            VK = t.text
            regex_rating = re.findall(r"(.+)\sRatings", VK)
            no_of_rating.append(regex_rating[0] if regex_rating else np.nan)

            regex_reviews = re.findall(r"&\s(.+)\sReviews", VK)
            no_of_reviews.append(regex_reviews[0] if regex_reviews else np.nan)
        else:
            no_of_rating.append(np.nan)
            no_of_reviews.append(np.nan)






    
    for i in soup.find_all("div",class_="KzDlHZ"):
        text=(i.text)

        regex = re.findall(r"^\w+",text)
        if regex:
            Brand.append(regex[0])
        else:
            Brand.append(np.nan)






     
    for i in soup.find_all("div",class_="KzDlHZ"):
        text=(i.text)

        regex = re.findall(r"(\s\d+\.?\d+\scm\s\([0-9]{1,}\sinch|Inch\))",text)
        if regex:
            Display.append(regex[0])
        else:
            Display.append(np.nan)


    for i in soup.find_all("div",class_="_6NESgJ"):
        text=i.text
        
        regex = re.findall(r"^\w+\s\w+:\s(\w+)",text)
        if regex:
            os.append(regex[0])
        else:
            os.append(np.nan)


    for i in soup.find_all("div",class_="_6NESgJ"):
        text=i.text
        
        regex = re.findall(r"^\w+\s\w+:\s(.+)\sPixelsLaunch",text)
        if regex:
            resolution.append(regex[0])
        else:
            resolution.append(np.nan)




    for i in soup.find_all("div",class_="_6NESgJ"):
        text=i.text
        
        regex = re.findall(r"PixelsLaunch\s\w+:\s(\d{4})",text)
        if regex:
            Lanch_Year.append(regex[0])
        else:
            Lanch_Year.append(np.nan)


    for i in soup.find_all("div",class_="_6NESgJ"):
        text=i.text
        
        regex = re.findall(r"PixelsLaunch\s\w+:\s\d{4}(\d\s\w+\s\w+)",text)
        if regex:
            Warranty.append(regex[0])
        else:
            Warranty.append(np.nan)




    for i in soup.find_all("div", class_="tUxRFH"):
        price_div = i.find("div", class_="Nx9bqj _4b5DiR")
        total.append(price_div.text if price_div else np.nan)




    for i in soup.find_all("div",class_="XQDdHH"):
        rq.append(i.text)
        

    



data=pd.DataFrame({"titles":titles,"Brand":Brand,"Display":Display,"os":os,"resolution":resolution,"Lanch_Year":Lanch_Year,"Warranty":Warranty,"Price":total,"Rating":Rating,"no_of_rating":no_of_rating,"no_of_reviews":no_of_reviews})


    





```

```python
import requests
import re
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup


Brand=[]
titles=[]
seater=[]
variants=[]
info=[]
Title=[]
city=[]
price=[]
delivery=[]
color=[]
save=[]
type=[]


for i in range(1,34):
    url="https://ackodrive.com/cars/page/{i}/"
    page=requests.get(url)
    soup=BeautifulSoup(page.text)



    for i in soup.find_all("div",class_="styles__ModelInfo-sc-66ca4afc-8 hngVXp"):
        titles.append(i.text)




    for i in soup.find_all("div",class_="styles__ModelInfo-sc-66ca4afc-8 hngVXp"):
        text=(i.text)

        regex = re.findall(r"^([a-zA-Z]+)",text)
        if regex:
            Brand.append(regex[0])
        else:
            Brand.append(np.nan)


    for i in soup.find_all("div",class_="styles__ModelInfo-sc-66ca4afc-8 hngVXp"):
        text=(i.text)
        
        regex = re.findall(r"(.)\sSeater\d.\d.",text)
        if regex:
            seater.append(regex[0])
        else:
            seater.append(np.nan)

    for i in soup.find_all("div",class_="styles__ModelInfo-sc-66ca4afc-8 hngVXp"):
        text=(i.text)
        
        regex = re.findall(r"Seater(\d.\d.)",text)
        if regex:
            variants.append(regex[0])
        else:
            variants.append(np.nan)


    for i in soup.find_all("div",class_="styles__Wrapper-sc-8077a65d-0 fZOnVd"):
        text=i.text
        if text:
            info.append(text)
        else:
            info.append(np.nan)
    
        regex = re.findall(r"^(\w+\s[A-Za-z]+\s?[a-zA-Z]+)",text)
        if regex:
            Title.append(regex[0])
        else:
            Title.append(np.nan)

        regex = re.findall(r"pricein\s(\w+)",text)
        if regex:
            city.append(regex[0])
        else:
            city.append(np.nan)

        regex = re.findall(r"pricein\s\w+(.\d+.\d+\s.)",text)
        if regex:
            price.append(regex[0])
        else:
            price.append(np.nan)


        regex = re.findall(r"pricein\s\w+.\d+.\d+\s.(.+days?)",text)
        if regex:
            delivery.append(regex[0])
        else:
            delivery.append(np.nan)

        regex = re.findall(r"\s(\d\scolors?)",text)
        if regex:
            color.append(regex[0])
        else:
            color.append(np.nan)

        regex = re.findall(r"days\s(.+)details?",text)
        if regex:
            save.append(regex[0])
        else:
            save.append(np.nan)


    for i in soup.find_all("div",class_="styles__VariantInfo-sc-66ca4afc-12 cjnSmx"):
         text=i.text
    
         regex = re.findall(r"^(\w+)A",text)
         if regex:
             type.append(regex[0])
         else:
             type.append(np.nan)





delhi=pd.DataFrame({"info":info,"Title":Title,"Brand":Brand,"seater":seater,"variants":variants,"city":city,"price":price,"delivery":delivery,"color":color,"save":save,"type":type})
delhi
    
```
